# ML lec 02 - Linear Regression의 Hypothesis 와 cost 설명

가설함수: 데이터의 분포를 설명하는 함수

cost funciton: 가설함수와 실제 데이터 값의 차이를 나타내는 함수

학습의 목표: cost function을 최소화 하는 파라미터를 찾는 것

```python
# placeholder를 사용하면 모델들을 재활용할 수 있다.

import tensorflow as tf

x_data = [1, 2, 3]
y_data = [1, 2, 3]


# Variable()과 placeholder()의 차이점이 무엇일까?
# Variable()은 가설함수의 파라미터
# placeholder()는 가설함수의 입력 및 출력값

# Variable: 모델이 업데이트가 가능한 값이 된다.
# y_data = W * x_data + b 를 계산하는 W와 b에 대한 값을 찾는다.
# (우리는 W가 1이고 b가 0임을 알고 잇지만, 텐서플로우는 그것을 이해하려고 할 것이다.)
W = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
b = tf.Variable(tf.random_uniform([1], -1.0, 1.0))

# 우리의 가설함수
hypothesis = W * x_data + b # 고정된 값을 주는 것이 아니라

# 단순화된 비용 함수
# 여기서 실제 계산이 일어는 나는 것은 아니다. 여기서는 operation만 정의하는 것이다.
cost = tf.reduce_mean(tf.square(hypothesis - y_data))

# 최소화한다. minimize하는 라이브러리를 사용한다.
a = tf.Variable(0.1)
optimizer = tf.train.GradientDescentOptimizer(a)
train = optimizer.minimize(cost)

# 변수들을 초기화하지 않고 실행하면, 초기화 하지 않았다는 에러 
# 메시지가 나올 것이다.
init = tf.initialize_all_variables()

# 그래프를 실행한다.
sess = tf.Session()
sess.run(init)

# Fit the line
for step in xrange(2001):
    sess.run(train)
    if step % 20 == 0:
        print step, sess.run(cost), sess.run(W), sess.run(b)
        
print(sess.run(hypothesis))
```

```python
# placeholder를 사용하면 모델들을 재활용할 수 있다.

import tensorflow as tf

x_data = [1, 2, 3]
y_data = [1, 2, 3]

# Variable: 모델이 업데이트가 가능한 값이 된다.
# y_data = W * x_data + b 를 계산하는 W와 b에 대한 값을 찾는다.
# (우리는 W가 1이고 b가 0임을 알고 잇지만, 텐서플로우는 그것을 이해하려고 할 것이다.)
W = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
b = tf.Variable(tf.random_uniform([1], -1.0, 1.0))

X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)


# 우리의 가설함수
# hypothesis = W * x_data + b # 고정된 값을 주는 것이 아니라
hypothesis = W * X + b # 모델을 실행시킬 때 값을 전달해준다.

# 단순화된 비용 함수
# 여기서 실제 계산이 일어는 나는 것은 아니다. 여기서는 operation만 정의하는 것이다.
# cost = tf.reduce_mean(tf.square(hypothesis - y_data))
cost = tf.reduce_mean(tf.square(hypothesis - Y))

# 최소화한다. minimize하는 라이브러리를 사용한다.
a = tf.Variable(0.1)
optimizer = tf.train.GradientDescentOptimizer(a)
train = optimizer.minimize(cost)

# 변수들을 초기화하지 않고 실행하면, 초기화 하지 않았다는 에러 
# 메시지가 나올 것이다.
init = tf.initialize_all_variables()

# 그래프를 실행한다.
sess = tf.Session()
sess.run(init)

# Fit the line
for step in xrange(2001):
    #sess.run(train)
    sess.run(train, feed_dict = {X: x_data, Y: y_data})
    if step % 20 == 0:
        #print step, sess.run(cost), sess.run(W), sess.run(b)
        print step, sess.run(cost, feed_dict={X: x_data, Y: y_data}), sess.run(W), sess.run(b)
        
# 모델을 다시 작성할 필요없이 placeholder X를 통해서 값을 전달할 수 있다.
print sess.run(hypothesis, feed_dict={X:5})
print sess.run(hypothesis, feed_dict={X:2.5})
```
